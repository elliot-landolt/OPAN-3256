#### Q38:

```python
df['log_gdpPercap'] = np.log(df['gdpPercap'])

scatter_plot = alt.Chart(df).mark_circle().encode(
    x='log_gdpPercap:Q',
    y='lifeExp:Q',
    tooltip=['country', 'log_gdpPercap', 'lifeExp'],
).properties(
    width=600,
    height=400
)

scatter_plot
```

Using the above graph as well as the calculation from #35 we can conclude that a higher gdp does correlate to higher life expectancy in most cases. this conclusion makes logical sense becuase higher GDP per capita typically means better access to healthcare, education, and other resources that contribute to improved life expectancy. Some countires might have additional influence due to the combination of both healthcare and governance. For the US we have strong healthcare however privatized which might cause life expectancy to be lower compared to countries with universal healthcare coverage.
